```python
# 6.Create a dataset and plot its histogram to visualize skewness
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
np.random.seed(34)
data=np.random.exponential(scale=2.0,size=1000);
plt.figure(figsize=(8,10));
sns.histplot(data,bins=30,kde=True,color='yellow');
plt.title('Skewness');
plt.xlabel('value');
plt.ylabel('Frequency');
plt.grid(True)
plt.show();
print(skew(data))


```

```python
#  8.Generate a dataset and demonstrate positive and negative skewness
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
np.random.seed(45)
right_skewd=np.random.exponential(scale=2.0,size=1000)
left_skewd= -right_skewd+max(right_skewd)+1;
plt.figure(figsize=(8,10));
sns.histplot(right_skewd,bins=30,kde=True,color='red');
plt.title("Right Skewed Data");
plt.xlabel("Value");
plt.ylabel("Frequency")
plt.grid(True);
plt.show();
sns.histplot(left_skewd,bins=30,kde=True,color='Skyblue');
plt.title("Left Skewed Data");
plt.xlabel("Value");
plt.ylabel("Frequency")
plt.grid(True);
plt.show();
print("the value of right Skewness: ",skew(right_skewd));
print("the value of left Skewness: ",skew(left_skewd));
```

```python
#  23.Visualize skewness and kurtosis using Python libraries like matplotlib or seaborn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew , kurtosis
np.random.seed(42);
data=np.random.randint(1,90,100);
data1=skew(data);
data2=kurtosis(data);
plt.figure(figsize=(8,10));
sns.histplot(data,bins=15,color="red",kde=True);
plt.title(f"Kurtosis: {data2:.2f}, Skewness: {data1:.2f}")
plt.xlabel("Data");
plt.ylabel("Frequancy");
plt.grid(True);
plt.show()

```

```python
# 19.Compare two datasets using their standard deviations
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(45);
data1=np.random.randint(1,20,5)
data2=np.random.randint(1,30,5)
std1=np.std(data1)
std2=np.std(data2);
plt.figure(figsize=(8,10));
sns.histplot(data1,bins=20,kde=True,color="red");
sns.histplot(data2,bins=20,kde=True,color="red");
plt.title(f"Standard1:{std1:.2f} , standard2: {std2:.2f}")
plt.xlabel("value")
plt.ylabel("frequency")
plt.grid(True)
plt.show()
```
